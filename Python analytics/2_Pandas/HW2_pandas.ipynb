### 4.4 Насколько в среднем стоимость домов с оценкой grade домов выше 10 отличается от стоимости домов с оценкой grade меньше 4?


```python
plus10 = df[df.grade > 10][['price']].mean()
minus4 = df[df.grade < 4][['price']].mean()
print("{:,.2f}".format(float(plus10 - minus4)))
```

```python
pd.io.formats.style.Styler((plus10 - minus4).rename('Difference'), precision=2, thousands=' ', caption='mean value')
```

### 5.2 Выберите дом клиенту
Клиент хочет дом либо с очень красивым видом из окна, либо с видом на набережную, в очень хорошем состоянии и год постройки не меньше 1980 года. В какой ценовом диапазоне будут дома?


```python
pd.DataFrame(df.waterfront.describe())
```

```python
sample2 = df[((df.waterfront != 0) | (df.view == 4)) & (df.condition == 5) & (df.yr_built >= 1980)]
sample2
```

```python
print("Дома будут в следующем ценовом диапазоне: ({:_.0f}, {:_.0f}).".format(min(sample2.price), max(sample2.price)).replace('_', ' '))
```
