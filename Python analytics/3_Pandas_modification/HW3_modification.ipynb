### 1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома
Если реновации дома не было, то в новом признаке поставьте 0

```python
df.info()
```

```python
def delta_f(x1, x2):
    return (x2 - x1) if (x2 != 0) else 0

df['delta_renovated'] = df.apply(lambda x: delta_f(x.yr_built, x.yr_renovated), axis=1)
pd.DataFrame(df[['id', 'yr_built', 'yr_renovated', 'delta_renovated']])
```

```python
pd.DataFrame(df.delta_renovated.value_counts())
```

### 1.3 Создайте признаки года продажи, месяца продажи


```python
df['yr_sold'] = df.date.str.slice(stop=4)
df['m_sold'] = df.date.str.slice(start=4, stop=6)
pd.DataFrame(df[['id', 'date', 'yr_sold', 'm_sold']])
```

```python
pd.DataFrame(df.yr_sold.value_counts())
```

```python
pd.DataFrame(df.m_sold.value_counts())
```

### 3.4 Каких домов в зависимости от этажности и количества спален больше?


```python
cross1 = pd.crosstab(index=df.floors, columns=df.bedrooms)
cross1
```

```python
rowmax1 = cross1.max(axis=1)
rowmax1[rowmax1 == max(rowmax1)]
```

```python
colmax1 = cross1.max(axis=0)
colmax1[colmax1 == max(colmax1)]
```

```python
print(f'Больше всего домов следующей этажности: {rowmax1[rowmax1 == max(rowmax1)].index.tolist()[0]}, \
и с количеством спален = {colmax1[colmax1 == max(colmax1)].index.tolist()[0]}.\
\nКоличество составляет: {rowmax1[rowmax1 == max(rowmax1)].tolist()[0]} домов.') 

# [0] - так как только 1 значение. Если было бы несколько, то надо думать, как реализовать
```
