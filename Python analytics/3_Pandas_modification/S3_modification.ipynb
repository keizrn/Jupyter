```python
def memory(x):
    dict1 = {'GB': 1,
            'TB': 1024,
            'MB': 0.0009765625}
    result = 0
    list1 = x.split(' +  ')
    for i in range(len(list1)):
        list_t = (list1[i]).split(' ')
        for k, v in dict1.items():
            if list_t[0].endswith(k):
                result += float(list_t[0].replace(k, '')) * v

    return int(result)
```

```python
df['Memory_Amount'] = df.Memory.apply(memory)
```

```python
def memory_type(x):
    result = ''
    list1 = x.split(' +  ')
    for i in range(len(list1)):
        list_t = (list1[i]).split(' ')
        result += '' if result == list_t[1] else (' + ' + list_t[1]) if result !='' else list_t[1]

    return result
```

```python
df['Memory_Type'] = df.Memory.apply(memory_type)
df
```

### 4.1 Найдите среднюю стоимость ноутбуков в зависимости от компании производителя
