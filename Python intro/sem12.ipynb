```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
x = df['Площадь, м^2']
y = df['Цена, тыс.руб,']
x
```

```python
y
```

```python
plt.scatter(x, y)
```

```python

sum_x = sum(x)
sum_y = sum(y)
sum_x2 = sum(pow(x, 2))
sum_xy = sum([x[i] * y[i] for i in range(len(x))])
leng = len(x)
print(sum_x, sum_y, sum_x2, sum_xy, sep='\n')
```

```python
a = (leng * sum_xy - sum_x * sum_y) / (leng * sum_x2 - sum_x**2)
a
```

```python
b = (sum_y - a * sum_x) / leng
b
```

```python
y2 = a * x + b
y2
```

```python
plt.scatter(x, y)
plt.plot(x, y2, 'red')
y3 = -0.05*x + 23
plt.plot(x, y3, 'green')
```
