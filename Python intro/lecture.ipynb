```python
import matplotlib as mpl
```

```python
x = list(range(-200, 200))
fx = [i**2 + 2 * i - 3 for i in x]
x2 = list(range(-200, 201, 25))
print(x2)
fx2 = [i2**2 + 2 * i2 - 3 for i2 in x2]
mpl.rcParams['lines.linewidth'] = 2
mpl.rcParams['lines.linestyle'] = '-.' #'--'
mpl.rcParams['lines.color'] = 'C0'

plt.axes(facecolor = 'teal')
plt.plot(x, fx, 'lime')
plt.plot(x2, fx2, 'wD')
plt.xlabel('x')
plt.ylabel('y = x + 2x - 3')
plt.show()
```

```python
vx = [1.0, 2.0, 3.0]
vy = [1.0, 2.0, 1.0]

sx = vx[0]
sy = vy[0]

px = []
py = []

for i in range(1, 1000):
    r = random.randint(0, 2)
    sx = (sx + vx[r]) / 2
    sy = (sy + vy[r]) / 2
    px.append(sx)
    py.append(sy)
    
plt.plot(px, py, 'o')
plt.show()
```

## Throw the dice

```python
count = 10000

x = list(range(1, 7))
y = [0 for i in x]

fig, ax = plt.subplots()

for _ in range(count):
    y[random.randint(0, 5)] += 1
    
ax.bar(x, y)
#
stat = [f'{i}: {y[i-1]}, {round(y[i-1] / count * 100, 2)}%' for i in x]
print(stat)
ax.set_xticks(x)
plt.show()
```
