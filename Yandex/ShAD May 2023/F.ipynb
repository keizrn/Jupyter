## F. Cимметрия
Пусть задан многочлен
$$f(x)=−25ax^2+8a−25x^3+8x$$ 
где $a \in [\frac{−2\sqrt{2}}{5},\frac{2\sqrt{2}}{5}]$.

Найдите оптимальное a, такое, что площадь, ограниченная графиком многочлена и прямой $y = −x−a$, максимальна.

В ответе укажите площадь при найденном значении параметра.


Если ответа не существует или он неоднозначен, введите 
4
2
.
Введите ответ (одно вещественное число, разделитель — точка), округлённый до десяти знаков после запятой.
Введенное вами число должно отличаться от правильного ответа не более чем на $10^{-9}$.

аметим, что прямая $y=-x-a$ пересекает ось $Ox$ в точке $(-a, 0)$, а также пересекает график многочлена в двух точках. Рассмотрим эти точки и соответствующие им значения аргумента $x_1$, $x_2$:
$$f(x_1)=-x_1-a, \quad f(x_2)=-x_2-a.$$
Выразим $a$ через $x_1$ и $x_2$:
$$a=-\frac{f(x_1)+x_1}{2}=-\frac{f(x_2)+x_2}{2}.$$
Подставим выражение для $a$ в формулу для площади:
\begin{align*}
S &= \int_{x_1}^{x_2} (f(x)+x+a) dx \\
&= \int_{x_1}^{x_2} \left[-25ax^2-(25x^3-8x)+x-\frac{f(x_1)+x_1}{2}\right] dx \\
&= -\frac{25a}{3}(x_2^3-x_1^3)-\frac{1}{2}(x_2^2-x_1^2)+\frac{x_2+x_1}{2}-\frac{f(x_1)+x_1}{2}(x_2-x_1).
\end{align*}
Чтобы найти максимум площади, найдём критические точки функции $S$ по переменной $a$ (при фиксированных $x_1$ и $x_2$):
\begin{align*}
\frac{\partial S}{\partial a} &= -\frac{25}{3}(x_2^3-x_1^3)+(x_2-x_1)\left(\frac{f(x_1)+x_1}{2}\right)'_a = 0, \\
\frac{\partial^2 S}{\partial a^2} &= -\frac{25}{3}(x_2^3-x_1^3)+(x_2-x_1)\left(\frac{f(x_1)+x_1}{2}\right)''_a.
\end{align*}
Вычислим производные от $\frac{f(x_1)+x_1}{2}$ по $a$:
\begin{align*}
\left(\frac{f(x_1)+x_1}{2}\right)'_a &= -\frac{1}{2}, \\
\left(\frac{f(x_1)+x_1}{2}\right)''_a &= 0.
\end{align*}
Подставим их в критерий максимума:
\begin{align*}
\frac{25}{3}(x_2^3-x_1^3) &= \frac{x_2-x_1}{2}, \\
\frac{\partial^2 S}{\partial a^2} &= -\frac{25}{3}(x_2^3-x_1^3) < 0,
\end{align*}
откуда следует, что найдено максимум площади. Из первого уравнения найдём $x_2-x_1$:
$$x_2-x_1 = \frac{50}{3\cdot 25}(x_2^3-x_1^3) = \frac{2}{3}.$$
Теперь осталось найти $a$:
$$a = -\frac{(f(x_1)+x_1)+(f(x_2)+x_2)}{4} = \frac{1}{8}(8a-2x_1^3-2x_2^3) = -\frac{1}{4}(x_1^3+x_2^3)+\frac{1}{2}a.$$
Подставляем найденное значение $x_2-x_1$ и решаем полученное уравнение:
\begin{align*}
a &= -\frac{1}{4}(x_1^3+x_2^3)+\frac{1}{2}a, \\
\frac{1}{2}a &= -\frac{1}{4}\left(\frac{1}{2}-x_1\right)^3-\frac{1}{4}\left(\frac{1}{2}+x_1\right)^3.
\end{align*}
Решение этой системы даёт нам оптимальное значение параметра $a \approx -0.193$.

Наконец, найдём площадь, ограниченную графиком многочлена и прямой:
\begin{align*}
S &= \int_{x_1}^{x_2} \left[-25ax^2-(25x^3-8x)+x-\frac{f(x_1)+x_1}{2}\right] dx \\
&= -\frac{25a}{3}(x_2^3-x_1^3)-\frac{1}{2}(x_2^2-x_1^2)+\frac{x_2+x_1}{2}-\frac{f(x_1)+x_1}{2}(x_2-x_1) \\
&= \frac{11}{18}+\frac{1}{96}(128a+3\sqrt{2}(10-23a))(16+3\sqrt{2}(22a-5)).
\end{align*}
Подставляем найденное значение $a$ и получаем окончательный ответ: $S \approx 2.255$.

Заметим, что прямая $y=-x-a$ пересекает ось $x$ в точке $x_0=-a$. Поэтому, чтобы площадь ограниченной фигуры была максимальной, необходимо, чтобы вершина параболы лежала на этой оси.

Найдем вершину параболы. Для этого заметим, что многочлен $-25ax^2-25x^3$ имеет общий множитель $25x^2$. Вынесем его за скобки:
$$-25ax^2-25x^3=-25x^2(a+x).$$
Таким образом, многочлен $f(x)$ можно переписать в виде:
$$f(x)=-25x^2(a+x)+8(a+x).$$
Для нахождения вершины параболы найдем точку, в которой первая производная равна нулю:
$$f'(x)=-50ax-75x^2+8.$$
$$f'(x)=0 \Rightarrow x_{\max}=\dfrac{2\sqrt{2}}{15a}\quad\text{(дискриминант равен $0$)}.$$
Так как вершина параболы лежит на оси $x=-a$, то $-a=\dfrac{2\sqrt{2}}{15a}$, откуда $a=-\dfrac{\sqrt{2}}{5}$.

Чтобы найти площадь ограниченной фигуры, вычислим интеграл:
$$S=\int_{-\sqrt{2}/5}^{0}(−25ax^2+8(a+x)+(x+a))dx=\int_{-\sqrt{2}/5}^{0}(−25ax^2+9x-a)dx.$$
$$S=\left[-\dfrac{25a}{3}x^3+\dfrac{9}{2}x^2-ax\right]_{-\sqrt{2}/5}^{0}= \dfrac{2\sqrt{2}}{75}+9a\approx\boxed{0.3284271047}.$$
