## Задача C
Пусть задан случайный вектор v = (X1, X2, X3), компоненты которого независимы и распределены равномерно на
отрезке [1, 5].

Найдите функцию плотности случайного вектора w = (X(3), X(2), X(1)), состоящего из компонент вектора v, упорядоченных по невозрастанию.

### Решение
Функция распределения случайного вектора $v$ будет равна:

$$
F_{v}(x_{1}, x_{2}, x_{3}) = P(X_{1} \leq x_{1}, X_{2} \leq x_{2}, X_{3} \leq x_{3}) = \left\{
\begin{aligned}
&0, &&\text{если}\ x_{1}, x_{2}, x_{3} < 1 \\
&\frac{(x_{1} - 1)(x_{2} - 1)(x_{3} - 1)}{64}, &&\text{если}\ 1 \leq x_{1}, x_{2}, x_{3} \leq 5 \\
&1, &&\text{если}\ x_{1}, x_{2}, x_{3} > 5 \\
\end{aligned}
\right.
$$

Для нахождения функции плотности распределения вектора $w$ нужно найти обратное отображение функции распределения $F_{v}(x_{1}, x_{2}, x_{3})$ относительно перестановки его компонент, упорядоченных по невозрастанию:

$$
w_{1} = X_{3}, \quad w_{2} = X_{2}, \quad w_{3} = X_{1}
$$

Таким образом, для перехода от вектора $v$ к вектору $w$ необходимо осуществить две перестановки компонент:

$$
v \to (X_{2}, X_{3}, X_{1}) \to w = (X_{3}, X_{2}, X_{1})
$$

Используя формулу замены переменных в многомерных интегралах, найдем функцию плотности распределения вектора $w$:

$$
f_{w}(x_{1}, x_{2}, x_{3}) = f_{v}(x_{3}, x_{2}, x_{1}) \left| J \right|
$$

где $J$ - матрица Якоби замены переменных:

$$
J = \begin{bmatrix}
0 & 0 & 1 \\
0 & 1 & 0 \\
1 & 0 & 0 \\
\end{bmatrix}
$$

Тогда:

$$
\begin{aligned}
f_{w}(x_{1}, x_{2}, x_{3}) &= f_{v}(x_{3}, x_{2}, x_{1}) \left| J \right| \\
&= f_{v}(x_{3}, x_{2}, x_{1}) \cdot 1 \\
&= f_{v}(X_{1}, X_{2}, X_{3}) \\
&= \frac{1}{64}, \quad 1 \leq x_{3} \leq x_{2} \leq x_{1} \leq 5 \\
&= 0, \quad \text{в остальных случаях}
\end{aligned}
$$

Таким образом, функция плотности случайного вектора $w$ равна $f_{w}(x_{1}, x_{2}, x_{3}) = \frac{1}{64}$, если выполнены условия $1 \leq x_{3} \leq x_{2} \leq x_{1} \leq 5$, и равна нулю во всех остальных случаях.
