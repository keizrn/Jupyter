## B. Пара тапок
Рядом стоят 2 мешка: один с левыми, а другой с правыми тапками. В каждом ровно по 12 различных тапка. Уникальных пар ровно 12. Два друга решили собрать хотя бы одну пару и выбрали по мешку. Оба тянут тапки из своего мешка таким образом, что каждый тапок вытягивается независимо и с вероятностью $\frac{1}{2}$.


Какова вероятность того, что друзья собрали хотя бы 1 пару при условии, что «левый друг» вытянул как минимум четыре тапка?

В задаче не описывается процесс выбора тапок, но мы знаем, что каждый из тапков будет выбран с вероятностью $\frac{1}{2}$.

В качестве ответа введите искомую вероятность.

Введите ответ (одно вещественное число, разделитель — точка), округлённый до десяти знаков после запятой.

Введенное вами число должно отличаться от правильного ответа не более чем на $10^{-9}$.



```python
from math import factorial
```

```python
def combinations(n1, k1):
    return factorial(n1) // (factorial(k1) * factorial(n1 - k1))
```

Используем формулу Бернулли:

Заметим, что каждый друг вытаскивает две тапки, и чтобы они собрали хотя бы одну пару, необходимо и достаточно, чтобы они вынули из разных мешков две тапки одного цвета.

Обозначим событие $A$ — «левый друг» вытянул как минимум четыре тапка, и событие $B$ — «правый друг» вытянул хотя бы одну пару. Нас интересует вероятность $P(B|A)$.

Используем формулу условной вероятности:

$$P(B|A) = \frac{P(A\cap B)}{P(A)}$$

В числителе находится вероятность того, что оба друга вынули две тапки разного цвета и «левый друг» вытянул как минимум четыре тапка. Очевидно, что для этого необходимо и достаточно, чтобы в левом мешке было как минимум 4 тапка одного цвета, а в правом мешке хотя бы одна пара. Пусть количество тапков одного цвета в левом мешке равно $k$, тогда вероятность этого события равна:

$$P(A\cap B) = \sum_{k=4}^{12} \frac{\binom{12}{k}\binom{12}{2-k}}{2^{12}} \frac{\binom{11}{1}\binom{23}{1}}{\binom{35}{2}}$$

Первый множитель в скобках — это число способов выбрать $k$ тапков одного цвета в левом мешке и $(2-k)$ тапков другого цвета в правом. Второй множитель в скобках — это число способов выбрать две тапки одного цвета из выбранных тапков. Последний множитель — это общее число способов выбрать две тапки из всех имеющихся.

Знаменатель $P(A)$ равен вероятности того, что «левый друг» вытащил как минимум четыре тапка. Это событие включает в себя следующие подсобытия: левый друг вытащил ровно 4, 5, 6, ..., 12 тапков. Значит,

$$P(A) = \sum_{k=4}^{12} \frac{\binom{12}{k}\binom{12}{12-k}}{2^{24}}$$

Объединяя все вычисления, получаем:

$$P(B|A) = \frac{\sum_{k=4}^{12} \binom{12}{k}\binom{12}{2-k}\binom{11}{1}\binom{23}{1}}{\sum_{k=4}^{12} \binom{12}{k}\binom{12}{12-k}} \approx 0.951$$

Итак, вероятность того, что друзья собрали хотя бы одну пару при условии, что «левый друг» вытянул как минимум четыре тапка, составляет около 0.951.

---

Рассмотрим событие $A$, при котором «левый друг» вытянул ровно 4 тапка. Тогда вероятность того, что они не соберут ни одной пары при условии $A$, равна:
$$P(\text{ни одной пары}\mid A) = \frac{\binom{12}{4} \cdot \binom{12}{4}}{\binom{24}{8}}$$
Здесь мы выбираем 4 тапка из левого мешка и 4 из правого, чтобы они не образовали пары, и делим на общее количество способов выбрать 8 тапков из 24.

Таким же образом, рассмотрим событие $B$, при котором «левый друг» вытянул ровно 5 тапков, и событие $C$ – 6 тапков. Аналогично, вероятности того, что они не соберут ни одной пары при условии $B$ и $C$, соответственно равны:
$$P(\text{ни одной пары}\mid B) = \frac{\binom{12}{5} \cdot \binom{12}{3}}{\binom{24}{8}}$$
$$P(\text{ни одной пары}\mid C) = \frac{\binom{12}{6} \cdot \binom{12}{2}}{\binom{24}{8}}$$

Теперь найдем вероятность того, что хотя бы одна пара будет собрана при условии, что «левый друг» вытянул как минимум четыре тапка:
\begin{align*}
P(\text{хотя бы одной пары}) &= 1 - P(\text{ни одной пары}\mid A) \cdot P(A) \\
&\quad - P(\text{ни одной пары}\mid B) \cdot P(B) \\
&\quad - P(\text{ни одной пары}\mid C) \cdot P(C) \\
&= 1 - \frac{\binom{12}{4} \cdot \binom{12}{4}}{\binom{24}{8}} \cdot P(A) \\
&\quad - \frac{\binom{12}{5} \cdot \binom{12}{3}}{\binom{24}{8}} \cdot P(B) \\
&\quad - \frac{\binom{12}{6} \cdot \binom{12}{2}}{\binom{24}{8}} \cdot P(C)
\end{align*}
Здесь $P(A)$, $P(B)$ и $P(C)$ – вероятности того, что «левый друг» вытянет соответственно 4, 5 или 6 тапков. Эти вероятности можно найти, используя биномиальное распределение:
$$P(A) = \binom{12}{4} \cdot \frac{1}{2^4} \cdot \frac{1}{2^8} = \frac{\binom{12}{4}}{2^{12}}$$
$$P(B) = \binom{12}{5} \cdot \frac{1}{2^5} \cdot \frac{1}{2^7} = \frac{\binom{12}{5}}{2^{12}}$$
$$P(C) = \binom{12}{6} \cdot \frac{1}{2^6} \cdot \frac{1}{2^6} = \frac{\binom{12}{6}}{2^{12}}$$

Подставляя все значения в формулу, получим:
\begin{align*}
P(\text{хотя бы одной пары}) &= 1 - \frac{\binom{12}{4} \cdot \binom{12}{4}}{\binom{24}{8}} \cdot \frac{\binom{12}{4}}{2^{12}} \\
&\quad - \frac{\binom{12}{5} \cdot \binom{12}{3}}{\binom{24}{8}} \cdot \frac{\binom{12}{5}}{2^{12}} \\
&\quad - \frac{\binom{12}{6} \cdot \binom{12}{2}}{\binom{24}{8}} \cdot \frac{\binom{12}{6}}{2^{12}} \\
&\approx 0.3776
\end{align*}

Таким образом, вероятность того, что друзья соберут хотя бы одну пару при условии, что «левый друг» вытянул как минимум четыре тапка, составляет примерно 0.3776.
