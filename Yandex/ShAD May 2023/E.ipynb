## E. Грам
Пусть задано евклидово пространство размерности 3. В нём выбрали систему из 3 векторов ранга 2. На этих векторах построили матрицу Грама, а затем умножили на −42.

Рассмотрим полученную матрицу как матрицу некоторой квадратичной формы на другом линейном пространстве над R в некотором фиксированном базисе.

При каком максимальном значении параметра c матрица

$$\begin{pmatrix}
-2c^2 - 12 & -c^2 - 9 & -6 \\
-c^2 - 9 & -c^2 - 9 & -c^2 - 9 \\
-6 & -c^2 - 9 & -2c^2 - 12 
\end{pmatrix}$$

не является матрицей той же формы, но в другом базисе?

Матрица Грама системы из трех векторов ранга 2 в евклидовом пространстве размерности 3 имеет вид:

$$G = \begin{pmatrix}
\langle v_1,v_1\rangle & \langle v_1,v_2\rangle &\langle v_1,v_3\rangle\\ 
\langle v_2,v_1\rangle & \langle v_2,v_2\rangle & \langle v_2,v_3\rangle \\ 
\langle v_3,v_1\rangle & \langle v_3,v_2\rangle & \langle v_3,v_3\rangle \\
\end{pmatrix}$$

Здесь $\langle a,b \rangle$ обозначает скалярное произведение векторов $a$ и $b$.

Так как ранг системы векторов равен 2, то её линейно зависимы. Пусть, например, $v_3$ можно выразить через $v_1$ и $v_2$:

$$v_3 = a_1v_1 + a_2v_2$$

Тогда матрица Грама системы имеет вид:

$$G = \begin{pmatrix}
\langle v_1,v_1\rangle & \langle v_1,v_2\rangle &a_1\langle v_1,v_1\rangle + a_2\langle v_1,v_2\rangle\\ 
\langle v_2,v_1\rangle & \langle v_2,v_2\rangle & a_1\langle v_1,v_2\rangle + a_2\langle v_2,v_2\rangle \\ 
a_1\langle v_1,v_1\rangle + a_2\langle v_1,v_2\rangle & a_1\langle v_1,v_2\rangle + a_2\langle v_2,v_2\rangle & a_1^2 \langle v_1,v_1\rangle + 2a_1a_2\langle v_1,v_2\rangle+ a_2^2\langle v_2,v_2\rangle \\
\end{pmatrix}$$

Так как ранг матрицы Грама равен 2 (из-за линейной зависимости системы векторов), то её определитель равен 0:

$$\det(G) = 0$$

Используя формулы для скалярного произведения векторов, можно составить систему уравнений на $a_1$ и $a_2$:

$$\begin{cases}
\langle v_1,v_1\rangle a_1 + \langle v_1,v_2\rangle a_2 + \langle v_1,v_3\rangle = 0\\
\langle v_1,v_2\rangle a_1 + \langle v_2,v_2\rangle a_2 + \langle v_2,v_3\rangle = 0
\end{cases}$$

Решая эту систему, получаем:

$$a_1 = \frac{\langle v_2,v_2\rangle \langle v_1,v_3\rangle - \langle v_1,v_2\rangle \langle v_2,v_3\rangle}{\Delta}$$

$$a_2 = \frac{\langle v_1,v_1\rangle \langle v_2,v_3\rangle - \langle v_1,v_2\rangle \langle v_1,v_3\rangle}{\Delta}$$

где $\Delta = \langle v_1,v_1\rangle \langle v_2,v_2\rangle - \langle v_1,v_2\rangle^2$.

Теперь можно выразить матрицу Грама через только два вектора $v_1$ и $v_2$. Пусть

$$G = \begin{pmatrix}
g_{11} & g_{12}\\ 
g_{12} & g_{22}\\ 
\end{pmatrix}$$

Тогда $g_{11} = \langle v_1,v_1\rangle$, $g_{12} = \langle v_1,v_2\rangle$, $g_{22} = \langle v_2,v_2\rangle$ и

$$\det(G) = \Delta = g_{11}g_{22} - g_{12}^2$$

Таким образом, матрица Грама системы из трех векторов можно выразить через только два вектора и их скалярное произведение. При этом, если система векторов линейно независима, то $\Delta \neq 0$ и тогда матрица Грама имеет полный ранг (равен 2). Если же система векторов линейно зависима, то $\Delta = 0$ и матрица Грама имеет ранг не больше 1.

Таким образом, матрица, заданная в условии, является матрицей квадратичной формы на некотором линейном пространстве над $\mathbb{R}$ в некотором фиксированном базисе, если и только если параметр $c$ таков, что система из трех векторов, построенная из столбцов этой матрицы, линейно зависима.

Выразим векторы этой системы:

$$v_1 = \begin{pmatrix}
-2c^2 - 12\\ 
-c^2 - 9\\ 
-6
\end{pmatrix}, \quad
v_2 = \begin{pmatrix}
-c^2 - 9\\ 
-c^2 - 9\\ 
-c^2 - 9
\end{pmatrix}, \quad
v_3 = \begin{pmatrix}
-6\\ 
-c^2 - 9\\ 
-2c^2 - 12
\end{pmatrix}$$

Составим матрицу из этих векторов:

$$A = \begin{pmatrix}
-2c^2 - 12 & -c^2 - 9 & -6 \\ 
-c^2 - 9 & -c^2 - 9 & -c^2 - 9 \\ 
-6 & -c^2 - 9 & -2c^2 - 12

Найдем определитель этой матрицы. Он равен:

$$\det(A) = -42c^6 - 216c^4 - 294c^2 - 108$$

Положим этот определитель равным нулю и решим полученное уравнение:

$$-42c^6 - 216c^4 - 294c^2 - 108 = 0$$

$$2c^6 + 12c^4 + 7c^2 + 3 = 0$$

При $c=\pm 1$ матрицы А и Грама совпадают (точнее, они отличаются знаком, но это не влияет на форму квадратичной функции), и, следовательно, имеются два базиса, в которых они одинаковы.
