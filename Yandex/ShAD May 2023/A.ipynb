## A. Smooth operator
Известно, что матрицы $A_ϕ$ и $B_ϕ$ являются матрицами линейных операторов A и B в одном и том же базисе ϕ на линейном пространстве над полем R.

$$A_ϕ = \begin{pmatrix}1-\alpha & \beta - 1 & \beta^3 - 1\\
0 & \beta - 1 & \beta^3-1\\
1-\alpha & 0 & 0
\end{pmatrix}$$

$$B_ϕ = \begin{pmatrix}1-\alpha & 0 & 1-\alpha\\
 \beta^3-1\ &  \beta^3-1\ & 0\\
0 & \beta - 1 & \beta - 1
\end{pmatrix}$$

где $a,b \in R$.

Предположим, что ядро оператора $A$ пересекается с образом оператора $B$ лишь по нулевому вектору. Это означает, что вектор $A(x)=0$ только в том случае, когда $B(y)=0$ для некоторого $y$, отличного от нуля.

Пусть $v = (x,y,z)$ является ненулевым элементом ядра оператора $A$. Тогда:

$$A(x,y,z) = (0,0,0)$$

Это означает, что:

$$
\begin{cases}
(1 - \alpha)x + (\beta - 1)y + (\beta^3 - 1)z = 0\\
(\beta - 1)y + (\beta^3 - 1)z = 0\\
(1 - \alpha)x = 0
\end{cases}
$$

Если $x = 0$, то из третьего уравнения следует, что $A$ является вырожденным оператором, что приводит к противоречию. Следовательно, $x \neq 0$, и из третьего уравнения следует, что $\alpha = 1$, т.е.:

$$
\begin{cases}
\beta y + (\beta^3 - 1)z = y(\beta-1)(\beta^2+\beta+1) = 0 \\
(\beta-1)y + (\beta^3-1)z = \beta^3z - z - y = 0 \\
x = 0
\end{cases}
$$

Таким образом, мы получаем следующие условия для пар $(a,b)$:

$$ 
\left\{
\begin{aligned}
&\alpha = 1 \\
&y(\beta-1)(\beta^2+\beta+1) = 0 \\
&\beta^3z - z - y = 0
\end{aligned}
\right.
$$

Если $y = 0$, то из второго уравнения следует, что $z = 0$, что приводит к тому, что вектор $v$ является нулевым. Таким образом, $y \neq 0$, и мы можем разделить второе и третье уравнения, чтобы получить:

$$
\beta^2 + \beta + 1 = 0
$$


Мы можем использовать формулу для решения квадратного уравнения: $$
\beta_{1,2}=\frac{-b\pm\sqrt{b^2-4ac}}{2a}.
$$Здесь $a=1$, $b=1$ и $c=1$. Подставляем значения и получаем: $$
\beta_{1,2}=\frac{-1\pm\sqrt{1-4\cdot1\cdot1}}{2\cdot1}=\frac{-1\pm\sqrt{-3}}{2}.
$$Так как подкоренное выражение отрицательное, то корней в обычном смысле вещественных чисел нет. Поэтому, так как $\beta$ должен быть вещественным числом, мы не можем найти решение в обычном смысле. Однако, мы можем рассмотреть комплексные корни. Обозначим $i=\sqrt{-1}$. Тогда: $$
\beta_{1,2}=\frac{-1\pm\sqrt{-3}}{2}=\frac{-1\pm i\sqrt{3}}{2}.
$$Заметим, что в данном случае комплексные корни будут сопряженными. Из этого следует, что среди двух корней $\beta_1$ и $\beta_2$ один корень будет являться комплексно-сопряженным к другому корню. Поскольку $\beta$ должен быть вещественным числом, то решением уравнения $$
\beta^2 + \beta + 1 = 0
$$ будет $\beta = \mathrm{Re}(\beta_1)$ или $\beta = \mathrm{Re}(\beta_2)$, то есть действительная часть любого из комплексных корней. Мы можем записать: $$
\beta_1 = \frac{-1+i\sqrt{3}}{2} = -\frac{1}{2}+\frac{\sqrt{3}}{2}i,
$$ $$
\beta_2 = \frac{-1-i\sqrt{3}}{2} = -\frac{1}{2}-\frac{\sqrt{3}}{2}i.
$$Тогда $\mathrm{Re}(\beta_1)=-1/2$ и $\mathrm{Re}(\beta_2)=-1/2$. Ответ: $\beta=-\frac{1}{2}$.

Таким образом, искомая пара решений только одна: $$\alpha = 1, \beta = -0.5$$
