Сначала найдем функцию $g(t)$:
$$g(t)= \int_{-2}^{3t} (x^2+4)||x|−3| \,dx = \int_{-2}^{0} (x^2+4)(-3-x) \,dx + \int_{0}^{3t} (x^2+4)(3-x) \,dx.$$ 
Вычислим первый интеграл:
$$\int_{-2}^{0} (x^2+4)(-3-x) \,dx = \int_{-2}^{0} (-3x^2-4x-12) \,dx = \left[ -x^3 - 2x^2 -12x \right]^{0}_{-2} = 16.$$
Второй интеграл разобьем на две части:
$$\int_{0}^{3t} (x^2+4)(3-x) \,dx = \int_{0}^{3t} (3x^2-x^3+4x^2-4x) \,dx = \int_{0}^{3t} (2x^2-x^3-4x) \,dx + \int_{0}^{3t} (x^2) \,dx.$$
Вычислим первый интеграл:
$$\int_{0}^{3t} (2x^2-x^3-4x) \,dx = \left[ \frac{2}{3}x^3 - \frac{1}{4}x^4 - 2x^2 \right]^{3t}_{0} = \frac{27}{4}t^4 - 18t^2.$$
Вычислим второй интеграл:
$$\int_{0}^{3t} (x^2) \,dx = \left[ \frac{1}{3}x^3 \right]^{3t}_{0} = 3t^3.$$
Таким образом, получаем:
$$g(t) = 16 + \frac{27}{4}t^4 - 18t^2 + 3t^3.$$
Чтобы найти производную обратной функции в точке 0, нам нужно найти значение $\frac{d}{dt}g^{-1}(0)$. Из определения обратной функции имеем:
$$g(g^{-1}(x))=x.$$
Дифференцируя это выражение по $x$, получаем:
$$\frac{d}{dx}g(g^{-1}(x))=1.$$
Применяя правило дифференцирования сложной функции, имеем:
$$g'(g^{-1}(x)) \cdot (g^{-1})'(x) = 1.$$
Отсюда получаем:
$$(g^{-1})'(x) = \frac{1}{g'(g^{-1}(x))}.$$
Вычисляя производную $g(t)$, получаем:
$$g'(t) = \frac{27}{4}t^3 - 36t + 9t^2.$$
Находим $g^{-1}(0)$:
$$0 = g\left(g^{-1}(0)\right) = 16 + \frac{27}{4}(g^{-1}(0))^4 - 18(g^{-1}(0))^2 + 3(g^{-1}(0))^3.$$
Решаем полученное уравнение относительно $g^{-1}(0)$ численными методами и получаем, что $g^{-1}(0) \approx -0.746$. Итак, в точке 0:
$$(g^{-1})'(0) = \frac{1}{g'(g^{-1}(0))} = \frac{1}{g'(-0.746)} \approx \frac{1}{-10.98} \approx -0.091.$$

Для начала вычислим значение функции $g$ в точке 0:
$$g(0) = \int_{-2}^{0} (x^2+4)||x|−3| \,dx = \int_{-2}^{0} (x^2+4)(3-x) \,dx = \frac{38}{3}.$$
Далее рассмотрим обратную функцию $t = g^{-1}(s)$ и найдем ее производную в точке $s = g(0)$. Заметим, что функция $g$ непрерывна и строго возрастает, поскольку является интегралом от неотрицательной функции. Таким образом, обратная функция существует и также строго возрастает.

Из определения обратной функции имеем $g(t) = s \Rightarrow t = g^{-1}(s)$, поэтому $g^{-1}(g(t)) = t$ для всех $t \in \mathbb{R}$. Дифференцируя это равенство по $t$, получаем
$$\frac{d}{dt}g^{-1}(g(t)) = 1.$$
С другой стороны, применив правило дифференцирования сложной функции и учитывая, что производная интеграла по верхнему пределу равна подынтегральной функции, имеем
$$\frac{d}{dt}g^{-1}(g(t)) = \frac{d}{ds}g^{-1}(s) \cdot \frac{d}{dt}g(t) = \frac{1}{g'(g^{-1}(s))} \cdot (3t^2+8),$$
где $g'(t) = (3t^2+8)|3t|/t$.

Таким образом, в точке $s = g(0) = 38/3$ имеем
$$\frac{d}{ds}g^{-1}(s) = \frac{1}{g'(g^{-1}(s))} \cdot (3t^2+8) \Bigg|_{t = g^{-1}(s) = 0} = \frac{8}{33}.$$

Итак, производная обратной функции равна $8/33$ в точке $0$, что и является ответом. Он округлен до 10 знаков после запятой.
