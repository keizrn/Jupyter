### Задача 3
Ниже приведены диаметры коронарных артерий после приема нифедипина и плацебо. Позволяют
ли приводимые ниже данные утверждать, что нифедипин влияет на диаметр коронарных артерий?
1. Выполнить расчеты в Python
2. Сделайте расчет в ручную
3. Сравните критерий Стьюдента и p-value со значениями, полученными в Python


x =np.array([2.5, 2.2, 2.6, 2, 2.1, 1.8,2.4, 2.3, 2.7, 2.7, 1.9])<br>
y= np.array([2.5, 1.7, 1.5, 2.5, 1.4, 1.9, 2.3, 2.0, 2.6, 2.3, 2.2])

### Решение
1. H<sub>0</sub>: Mx = My (разницы между группами нет)<br>
H<sub>1</sub>: Mx <> My
2. Alpha = 0.05 (уровень значимости 5%)
3. 
Формула, когда две выборки и сравниваем среднее:

Внимательно!! Формула ниже используется, когда одна выборка и проверяем её против константы!!


```python
import numpy as np
import math
from scipy import stats
```

```python
x =np.array([2.5, 2.2, 2.6, 2, 2.1, 1.8,2.4, 2.3, 2.7, 2.7, 1.9])
y= np.array([2.5, 1.7, 1.5, 2.5, 1.4, 1.9, 2.3, 2.0, 2.6, 2.3, 2.2])
print(x, y, sep='\n')
```

```python
Evx = np.mean(x)
Evy = np.mean(y)
Varx = np.var(x, ddof=1)
Vary = np.var(y, ddof=1)
lenx = len(x)
leny = len(y)
Evx, Evy, Varx, Vary, lenx, leny
```

```python
t_3 = (Evx - Evy) / math.sqrt(Varx/lenx + Vary/leny)
t_3 # t наблюдаемое
```

```python
alpha_3 = 0.05
t_3cr_1 = stats.t.ppf(alpha_3/2, df=(lenx + leny - 2))
t_3cr_2 = stats.t.ppf(1 - alpha_3/2, df=(lenx + leny - 2))
t_3cr_1, t_3cr_2 # t критические
```

```python
print("Гипотеза принимается") if t_3cr_1 < t_3 < t_3cr_2 else print("Гипотеза не принимается")
```

Проверяем через встроенные функции.

Тест двухвыборочный, независимый.

```python
stat_3, pvalue_3 = stats.ttest_ind(x, y)
stats.ttest_ind(x, y)
```

```python
print("Гипотеза принимается") if pvalue_3 > alpha_3 else print("Гипотеза не принимается")
```

### Задача 4
Ниже приведены данные из исследования Фреба и Уайта, посвященному исследованию
состоянию легких. . Мы возьмем данные для группы людей, которые работали в накуренном
помещении и для людей, выкуривающих небольшое число сигарет в день.

Объемы выборок одинаковые – по 200 человек. Для людей, работающих в накуренном
помещении средняя скорость средины выдоха составляет 2,72, std = 0.71, а выкуривающих
небольшое число сигарет 2,63, std = 0.73.Можно ли считать одинаковой максимальную
объемную скорость середины выдоха одинаковой в обеих группах?

```python
Evx_4 = 2.72
Evy_4 = 2.63
Varx_4 = 0.71**2
Vary_4 = 0.73**2
lenx_4 = 200
leny_4 = 200
print(Evx_4, Evy_4, Varx_4, Vary_4, lenx_4, leny_4, sep='\n')
```

```python
t_4 = (Evx_4 - Evy_4) / math.sqrt(Varx_4/lenx_4 + Vary_4/leny_4)
t_4
```

```python
alpha_4 = 0.05
t_4cr_1 = stats.t.ppf(alpha_4/2, df = (lenx_4 + leny_4 - 2))
t_4cr_2 = stats.t.ppf(1 - alpha_4/2, df = (lenx_4 + leny_4 - 2))
t_4cr_1, t_4cr_2
```

```python
print("Гипотеза принимается") if t_4cr_1 < t_4 < t_4cr_2 else print("Гипотеза не принимается")
```

арифметическое 178,5. Проверить
это утверждение с помощью одностороннего
теста, если известно, что стандартное
отклонение генеральной совокупности 3 см.
А уровень статистической значимости
принять за 1%
### Решение
1. H<sub>0</sub>: Mx = 179,5<br>
H<sub>1</sub>: Mx < 179,5
2. Alpha = 0.01 (уровень значимости 1%)
3. 



```python
Ev5 = 178.5
Sd5 = 3
len5 = 100
alpha5 = 0.01
z_5 = (Ev5 - 179.5) / (Sd5 / math.sqrt(len5))
z_5
```

```python
z_5cr = stats.norm.ppf(alpha5) # левосторонний тест
z_5cr
```

```python
print("Гипотеза принимается") if z_5 > z_5cr else print("Гипотеза не принимается")
```
