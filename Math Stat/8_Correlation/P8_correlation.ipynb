### Задача 2
Провести двусторонний тест и ответить на вопрос, есть ли статистически значимые
различия между средними 2х нормально распределенных генеральных совокупностей,
представленных следующими независимыми выборками:<br>
 a = np.array([12, 10, 11, 19, 13, 11, 17, 15, 19, 14, 21, 18, 21, 11, 17, 14, 15, 17, 20, 19])<br>
 b = np.array([11, 13, 18, 15, 17, 18, 10, 21, 26, 15, 11, 12, 15, 17, 10, 18, 18, 12,21, 20])
 
Уровень статистической значимости принять за 5%
1. Используйте функцию в Python.
2. Имея p-value из функции рассчитать наблюдаемое значение критерия.


H0: mu = mu1 = mu2

H1: mu1 <> mu2

```python
import math
```

```python
a_1 = np.array([12, 10, 11, 19, 13, 11, 17, 15, 19, 14, 21, 18, 21, 11, 17, 14, 15, 17, 20, 19])
b_1 = np.array([11, 13, 18, 15, 17, 18, 10, 21, 26, 15, 11, 12, 15, 17, 10, 18, 18, 12,21, 20])
print(a_1, b_1, sep='\n')
```

```python
mu1 = np.mean(a_1)
mu2 = np.mean(b_1)
var1 = np.var(a_1, ddof=1)
var2 = np.var(b_1, ddof=1)
len1 = len(a_1)
len2 = len(b_1)
mu1, mu2, var1, var2, len1, len2
```

```python
print(stats.ttest_ind(a_1, b_1, equal_var=True))
```

```python
t_1 = (mu1 - mu2) / (math.sqrt(var1/len1 + var2/len2))
t_1
```

```python
alpha1 = 0.05
tcr1_1 = stats.t.ppf(alpha1/2, df=len1 + len2 - 2)
tcr1_2 = stats.t.ppf(1 - alpha1/2, df=len1 + len2 - 2)
tcr1_1, tcr1_2
```

```python
print("Гипотеза принимается") if tcr1_1 < t_1 < tcr1_2 else print("Гипотеза отвергается")
```

```python
t_1x, p_1 = stats.ttest_ind(a_1, b_1, equal_var=True)
print("Гипотеза принимается") if p_1 > alpha1 else print("Гипотеза отвергается")
```

```python
t_1xx = stats.t.ppf(p_1/2, df=len1 + len2 - 2)
t_1xx
```
