```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
s = np.array([27, 37, 42 , 48, 57, 56, 77, 80])
p = np.array([1.2, 1.6, 1.8, 1.8, 2.5, 2.6, 3, 3.3])
```

```python
plt.scatter(s, p)
```

```python
cov = np.mean(p*s) - np.mean(p)*np.mean(s)
cov
```

 :

```python
np.cov(p, s) #   - ,  
```

```python
np.var(p, ddof=1), np.var(s, ddof=1) #      (!)
```

```python
np.corrcoef(p, s)
```

```python
stats.spearmanr(p, s)
```
