## Combinations

```python
from math import factorial
```

```python
def combinations(n, k):
    return factorial(n) // (factorial(k) * factorial(n - k))
combinations(36, 4)
```

## Arrangements:

```python
def arrangements(n2, k2):
    return factorial(n2) // factorial(n2 - k2)
arrangements(20, 5)
```

## Permutations:

```python
def permutations(n3):
    return factorial(n3)
permutations(5)
```
