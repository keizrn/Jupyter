## Random variables
Случайная величина — переменная, значения которой представляют собой численные исходы некоторого случайного феномена или эксперимента.

Другими словами, это численное выражение результата случайного события.



### Probabilty density function
Закон распределения вероятностей случайной величины - это соответствие между возможными значениями случайной величины и вероятностями, которые этим значениям соответствуют.

### Задача 1
3 раза подбрасываем монетку (n=3). Какая вероятность, что  орел выпадет 0 раз, 1 раз, 2 раза и 3 раза?


```python
from math import factorial
```

```python
def combinations(n1, k1):
    return factorial(n1) // (factorial(k1) * factorial(n1 - k1))
```

```python
p_h0 = combinations(3, 0) * 0.5**0 * 0.5**3
p_h0
```

```python
p_h1 = combinations(3, 1) * 0.5**1 * 0.5**2
p_h1
```

```python
p_h2 = combinations(3, 2) * 0.5**2 * 0.5**1
p_h2
```

```python
p_h3 = combinations(3, 3) * 0.5**3 * 0.5**0
p_h3
```

### Задача 2
Вероятность того, что среди писем, поступающих на определенный почтовый ящик, встретится письмо со спамом, составляет 0.01. Найдите вероятность того, что среди 1000 писем, поступивших на него за месяц, будет 11 со спамом.

```python
from math import exp
```

```python
k_2 = 11
n_2 = 1000
lambda_2 = 0.01 * n_2
p_11 = lambda_2**k_2 * exp(1)**(-lambda_2) / factorial(k_2)
p_11
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
x_2 = np.arange(0, 20)
def poisson_0(x_0):
    return lambda_2**x_0 * exp(1)**(-lambda_2) / factorial(x_0)
fx_2 = [poisson_0(i) for i in x_2]
```

```python
plt.figure(figsize=(5, 2.7), layout='constrained')
mpl.rcParams['lines.color'] = 'C0'
plt.plot(x_2, fx_2, label='spam in 1000')
plt.plot(x_2, fx_2, 'mD')
plt.xlabel('spam letters per 1000')
plt.ylabel('probability')
plt.title("Probability of spam in 1000 letters")
```
