### Задача 1
Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.

### Решение:
В решении используем формулу Бернулли:

```python
from math import factorial
from math import exp
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
def combinations(n1, k1):
    return factorial(n1) // (factorial(k1) * factorial(n1 - k1))
```

```python
p_85 = combinations(100, 85) * 0.8**85 * 0.2**15
print(f"Ответ: вероятность, что стрелок попадет в цель ровно 85 раз из 100: {round(p_85*100, 2)} %")
```

```python
x_1 = np.arange(0, 100)
def shooter_1(x_0):
    return combinations(100, x_0) * 0.8**x_0 * 0.2**(100 - x_0)
fx_1 = [shooter_1(i) for i in x_1]
```

```python
plt.figure(figsize=(5, 2.7), layout='constrained')
mpl.rcParams['lines.color'] = 'C0'
plt.plot(x_1, fx_1, label='shooter')
plt.xlabel('')
plt.ylabel('')
plt.title("Shooter")
```

### Задача 2
Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. Какова вероятность, что ни одна из них не перегорит в первый день? Какова вероятность, что перегорят ровно две?

### Решение:
В решении используем распределение Пуассона:

```python
k_2a = 0
k_2b = 2
n_2 = 5000
lambda_2 = 0.0004 * n_2
p_2a = lambda_2**k_2a * exp(1)**(-lambda_2) / factorial(k_2a)
p_2b = lambda_2**k_2b * exp(1)**(-lambda_2) / factorial(k_2b)
print(f"Ответ: вероятность, вероятность, что ни одна из лампочек не перегорит в первый день: {round(p_2a*100, 2)} %")
print(f"Ответ: вероятность, вероятность, что две лампочки перегорят в первый день: {round(p_2b*100, 2)} %")
```

```python
x_2 = np.arange(0, 10)
def bulb_2(k_0):
    return lambda_2**k_0 * exp(1)**(-lambda_2) / factorial(k_0)
fx_2 = [bulb_2(i) for i in x_2]
```

```python
plt.figure(figsize=(5, 2.7), layout='constrained')
mpl.rcParams['lines.color'] = 'C2'
plt.plot(x_2, fx_2, 'mD')
plt.xlabel('')
plt.ylabel('')
plt.title("Bulb")
```

### Задача 3
Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?
### Решение:
В решении используем формулу Бернулли:

```python
p_3 = combinations(144, 70) * 0.5**70 * 0.5**(144-70)
print(f"Ответ: вероятность, что орел выпадет ровно 70 раз: {round(p_3*100, 2)} %")
```

```python
x_3 = np.arange(0, 144)
def coin_3(x_0):
    return combinations(144, x_0) * 0.5**x_0 * 0.5**(144-x_0)
fx_3 = [coin_3(i) for i in x_3]
```

```python
plt.figure(figsize=(5, 2.7), layout='constrained')
mpl.rcParams['lines.color'] = 'C2'
plt.plot(x_3, fx_3)
plt.xlabel('')
plt.ylabel('')
plt.title("Coin 144 times")
```
