## Наибольшая вероятность


### Задача 1
Подбрасывают 4 одинаковые монеты. Какова вероятность, что решка выпадет
не более 1 раза

### Решение
Используем формулу Бернулли:

```python
from math import factorial
from math import exp
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
def combinations(n1, k1):
    return factorial(n1) // (factorial(k1) * factorial(n1 - k1))
```

```python
p_0 = combinations(4, 0) * 0.5**0 * 0.5**4
p_0
```

```python
p_1 = combinations(4, 1) * 0.5**1 * 0.5**3
p_1
```

```python
p_res = p_0 + p_1
p_res
```

### Задача 2

выпаданий тройки было 30. 
### Решение
Используем наибольшую вероятность

```python
k_cub = 30
p_2_3 = 1/6
q_2_3 = 5/6
n_2_1 = (k_cub + q_2_3)//p_2_3
n_2_2 = (k_cub - p_2_3)//p_2_3
print(n_2_1, n_2_2)
```

```python
for n_2 in range(int(n_2_2), int(n_2_1)+1):
    print(n_2)
```
